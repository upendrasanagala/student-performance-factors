# Data Visualization

### countplot for categorical data 

```python
# Select only the categorical columns (dtype = 'object')
categorical_cols = student.select_dtypes(include='object').columns
# Loop through each categorical column
for col in categorical_cols:
    # Create a count plot for each category in the column
    sns.countplot(x=col, data=df_cleaned)
    # Rotate x-axis labels for better readability (especially for long category names)
    plt.xticks(rotation=45)
    # Set the plot title dynamically using the column name
    plt.title(f'Count of {col}')
    # Adjust layout to prevent clipping of labels/titles
    plt.tight_layout()
    plt.show()
```

```python
# Loop through all columns in the DataFrame
for col in student:
    # Check if the column is not categorical (i.e., numeric type)
    if student[col].dtype != 'O':  #  'O' stands for object
        sns.histplot(student[col],kde=True) 
        #Plot histogram with Kernel Density Estimate for numeric column  
        plt.show()
```

### Family Income vs Exam Score

```python
sns.violinplot(x='Family_Income', y='Exam_Score', data=student)
# violinplot combines a box plot (median, quartiles, outliers) .
# A kernel density plot (distribution shape)
plt.title('Family Income vs Exam Score')
plt.show()
```

### Major factors vs Exam score

```python
selected_columns = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Sleep_Hours', 'Tutoring_Sessions', 'Exam_Score']
# Create pairwise scatter plots + histograms for each combination of selected features
sns.pairplot(student[selected_columns])
# Add a main title slightly above the top of the grid
plt.suptitle('Pair Plot of Selected Variables', y=1.02)
plt.show()
```

### Boys vs Girls

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Exam_Score', data=df_cleaned, hue='Gender', palette='Set2', legend=False)
plt.title('Exam Score Comparison: Boys vs Girls')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.tight_layout()
plt.show()
```

### Correlation Matrix of All Variables

- Converts text categories into numbers so they can be used in correlation or machine learning.

```python
# Selecting numerical features
numerical_cols = ['Hours_Studied', 'Attendance', 'Sleep_Hours',
                  'Previous_Scores', 'Tutoring_Sessions', 'Physical_Activity', 'Exam_Score']

# Compute and visualize correlation matrix
corr = df_cleaned[numerical_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```
